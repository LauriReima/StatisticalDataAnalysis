```python
import pandas as pd
import numpy as np
```

```python
columns = []
with open('allbp.names', 'r') as name_file:
    lines = name_file.readlines()
    for line in lines:
        if ':' in line:
            column_name = line.split(':')[0].strip()
            columns.append(column_name)
columns.append('classes')
```

```python
data = pd.read_csv('allbp.data', na_values=['?'],names=columns)
```
