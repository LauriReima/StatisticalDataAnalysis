```python
def normal_dist(df):
    count = 0
    bottom = df.mean() - df.std()
    top = df.mean() + df.std()
    for i in df:
        if bottom < i and i < top:
            count += 1
    return count/len(df)
```

```python
normal_dist(data1)
```
