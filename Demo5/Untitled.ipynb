An advertisement company followed online customers to discover how effectively advertisements lure them to spend money. The data files 54-image.csv and 54-video.csv contain information on how much customers spent in total after clicking on advertisements.

Is there statistical evidence to claim that the total amount spent by customers is different if they click on image advertisements than on video advertisements?

```python
import pandas as pd
from scipy import stats
```

```python
image = pd.read_csv('Data files-20230929/54-image.csv')
video = pd.read_csv('Data files-20230929/54-video.csv')
```

```python
img = image.rename(columns={'amount_spent': 'image_spent'})
vid = video.rename(columns={'amount_spent': 'video_spent'})
```

```python
result = pd.concat([img, vid], axis=0)
```

```python
print('image: total_spent: ',img['image_spent'].sum(), 'out of ', img['image_spent'].count(), '.mean: ',img['image_spent'].mean())
print('video: total_spent: ',vid['video_spent'].sum(), 'out of ', vid['video_spent'].count(), '.mean: ',vid['video_spent'].mean())
```

```python
AP = scipy.stats.shapiro(img['image_spent'])
BP = scipy.stats.shapiro(vid['video_spent'])
```
