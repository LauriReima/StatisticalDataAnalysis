```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('Data files-20230929/53-data.csv')
week_data = data[['Eteläesplanadi', 'Kaivokatu','Kuusisaarentie',	'Merikannontie',	'Länsituulenkuja', 'weekday']]
```

```python
week_cyc = week_data.groupby([ "weekday"]).mean()
week_cyc
```

```python
week_cyc.plot(kind="bar", figsize=(10, 6))
```
