Download and unzip the DataSet SOFT file from https://www.ncbi.nlm.nih.gov/sites/GDSbrowser?acc=GDS5037 . The SOFT file format contains tabular gene expression data accompanied by its metadata. The rows that start with either ^, !, or # are metadata, which can be ignored in this exercise.

Some useful pieces of information about the file GDS5037.soft:

The first 160 rows and the last row are metadata.
The ID_REF column contains probe ids. Use them as row names.
The IDENTIFIER column contains gene names.
The other columns contain gene expression levels for subjects.
Load and filter the data such that your data frame only contains the probe and subject ids listed in the files 51-probes.ids and 51-subjects.ids, respectively.

a. For each subject, calculate the mean expression level over the genes. Do you notice any obvious differences between the subjects?

b. For each probe, calculate the mean expression level over the subjects. Do you notice any obvious differences between the probes?

c. How many probes are there with a mean greater than 10.0? Which genes do they correspond to?

```python
data = pd.read_csv('GDS5037.soft' ,sep='\t',index_col=0,skiprows=160,skipfooter=1, engine='python' )
```

```python
columns = []
rows = []
with open('Data files-20230929/51-probes.ids', 'r') as file1:
    for line in file1:
        name1 = line.strip()
        columns.append(name1)
with open('Data files-20230929/51-subjects.ids', 'r') as file2:
    for line in file2:
        name2 = line.strip()
        rows.append(name2)
```

```python
filtered_df = data[data.index.isin(columns)]
```

```python
fd = filtered_df[rows]
```

```python
fd
```

```python
# a)
subject_mean = fd.mean(axis=0)
```

```python
print(subject_mean)
print('min value: ',subject_mean.min(), 'max value: ', subject_mean.max())
```

There aren't any obvious differences between subject means

```python
# b) 
probe_mean = fd.mean(axis=1)
```

```python
print(probe_mean)
print('min value: ',probe_mean.min(), 'max value: ', probe_mean.max())
```

There are big differences in probe means

```python
plusten = probe_mean[probe_mean > 10]
len(plusten)
```

```python
for id in plusten.index:
    print(data['IDENTIFIER'].loc[id])
```
